```python
df = pd.read_csv('cleaned_stock_data.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
```

```python
df.dtypes
```

```python
columns_to_clean = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']

for col in columns_to_clean:
    df[col] = df[col].str.replace(',', '').astype(float)
```

```python
df.dtypes
```

```python
df['SMA_20'] = df['Close'].rolling(window=20).mean()
df['EMA_20'] = df['Close'].ewm(span=20, adjust=False).mean()
```

```python
delta = df['Close'].diff()
gain = delta.clip(lower=0)
loss = -delta.clip(upper=0)
```

```python
avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()
```

```python
rs = avg_gain / avg_loss
df['RSI_14'] = 100 - (100 / (1 + rs))
```

```python
df['BB_Middle'] = df['Close'].rolling(window=20).mean()
df['BB_Std'] = df['Close'].rolling(window=20).std()

df['BB_Upper'] = df['BB_Middle'] + 2 * df['BB_Std']
df['BB_Lower'] = df['BB_Middle'] - 2 * df['BB_Std']
```

```python
df
```

```python
df.to_csv("stock_with_indicators.csv", index=False)
```

```python
from sklearn.ensemble import IsolationForest
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("stock_with_indicators.csv")
df_model = df.dropna().copy()
```

```python
features = ['Close', 'SMA_20', 'EMA_20', 'RSI_14', 'BB_Upper', 'BB_Lower']
X = df_model[features]
```

```python
model = IsolationForest(contamination=0.01, random_state=42)  # ~1% anomalies
df_model['Anomaly'] = model.fit_predict(X)
```

```python
df_model['Anomaly'] = df_model['Anomaly'].map({1: 0, -1: 1})
df_model.to_csv("anomaly_detected.csv", index=False)
print("Total anomalies found:", df_model['Anomaly'].sum())
```

```python
plt.figure(figsize=(14, 6))
plt.plot(df_model['Date'], df_model['Close'], label='Close Price', color='blue')

plt.title('Stock Price with Anomalies')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
anomalies = df_model[df_model['Anomaly'] == 1]
plt.scatter(anomalies['Date'], anomalies['Close'], color='red', label='Anomalies', s=50)
plt.title('Stock Price with Anomalies')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from prophet import Prophet
```

```python
df = pd.read_csv("stock_with_indicators.csv")
df_prophet = df[['Date', 'Close']].copy()
df_prophet.rename(columns={'Date': 'ds', 'Close': 'y'}, inplace=True)
```

```python
df_prophet.isnull().sum()
```

```python
model = Prophet(daily_seasonality=True)
model.fit(df_prophet)

future = model.make_future_dataframe(periods=60)
forecast = model.predict(future)
forecast
```

```python
model.plot(forecast)
plt.title("Stock Price Forecast with Prophet")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
df_actual = pd.read_csv("stock_with_indicators.csv")[['Date', 'Close']]
df_actual.rename(columns={'Date': 'ds', 'Close': 'y'}, inplace=True)

df_actual['ds'] = pd.to_datetime(df_actual['ds'])

merged = df_actual.merge(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']], on='ds')

merged['Forecast_Anomaly'] = ((merged['y'] > merged['yhat_upper']) | (merged['y'] < merged['yhat_lower'])).astype(int)
merged.to_csv("forecast_anomalies.csv", index=False)
print("Forecast-based anomalies found:", merged['Forecast_Anomaly'].sum())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,6))
plt.plot(merged['ds'], merged['y'], label='Actual Price', color='blue')
plt.plot(merged['ds'], merged['yhat'], label='Predicted Price (yhat)', color='green')

anomaly_points = merged[merged['Forecast_Anomaly'] == 1]
plt.scatter(anomaly_points['ds'], anomaly_points['y'], color='red', label='Forecast Anomaly', s=50)

plt.title("Forecast vs Actual with Anomalies")
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
isolation_df = pd.read_csv("anomaly_detected.csv")[['Date', 'Close', 'Anomaly']]
prophet_df = pd.read_csv("forecast_anomalies.csv")[['ds', 'y', 'Forecast_Anomaly']]
```

```python
prophet_df.rename(columns={'ds': 'Date', 'y': 'Close'}, inplace=True)
isolation_df['Date'] = pd.to_datetime(isolation_df['Date'])
prophet_df['Date'] = pd.to_datetime(prophet_df['Date'])
```

```python
combined = pd.merge(isolation_df, prophet_df, on=['Date', 'Close'], how='inner')
combined['Combined_Anomaly'] = ((combined['Anomaly'] == 1) | (combined['Forecast_Anomaly'] == 1)).astype(int)

combined.to_csv("combined_anomalies.csv", index=False)
print("Total combined anomalies:", combined['Combined_Anomaly'].sum())
```

```python
plt.figure(figsize=(14,6))
plt.plot(combined['Date'], combined['Close'], label='Stock Price', color='gray')

iso = combined[combined['Anomaly'] == 1]
plt.scatter(iso['Date'], iso['Close'], color='orange', label='Isolation Forest', s=50)

prophet = combined[combined['Forecast_Anomaly'] == 1]
plt.scatter(prophet['Date'], prophet['Close'], color='red', label='Prophet Forecast', s=30)

plt.title("Stock Price with Anomalies Detected")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
